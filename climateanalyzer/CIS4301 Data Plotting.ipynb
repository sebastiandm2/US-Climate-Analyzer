```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.dates import drange 

data = pd.read_csv("US Climate Normal Data.csv") 

data
```

```python
city_data= data[data.City == 'Miami']
city_data
```

```python
city_data["dt"]=pd.to_datetime(city_data["dt"])
city_data= city_data.sort_values(["dt"])
pd.set_option("max_rows", None)
city_data
```

```python
fig, ax = plt.subplots()
ax.plot_date(city_data.dt, city_data.AverageTemperature, label="City data")

ax.set_title('Average daily temperature for Specified City')
plt.legend(loc='best')
plt.savefig('miami_data.png')
plt.show()
```
