```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.dates import drange 
from matplotlib.font_manager import FontProperties


data = pd.read_csv("US Climate Normal Data.csv") 

data
```

```python
cityName="Miami"
city_data= data[data.City == cityName]
city_data
```

```python
city_data["dt"]=pd.to_datetime(city_data["dt"])
city_data= city_data.sort_values(["dt"])
pd.set_option("max_rows", None)
city_data
```

```python
fig, ax = plt.subplots()
fontP=FontProperties()
fontP.set_size('small')

pres_Roosevelt = city_data[city_data["President"]=='Roosevelt']
ax.scatter(pres_Roosevelt.dt, pres_Roosevelt.AverageTemperature, label='Roosevelt')
pres_Truman = city_data[city_data["President"]=='Truman']
ax.scatter(pres_Truman.dt, pres_Truman.AverageTemperature, label='Truman')
pres_Eisenhower = city_data[city_data["President"]=='Eisenhower']
ax.scatter(pres_Eisenhower.dt, pres_Eisenhower.AverageTemperature, label='Eisenhower')
pres_Kennedy = city_data[city_data["President"]=='Kennedy']
ax.scatter(pres_Kennedy.dt, pres_Kennedy.AverageTemperature, label='Kennedy')
pres_Johnson = city_data[city_data["President"]=='Johnson']
ax.scatter(pres_Johnson.dt, pres_Johnson.AverageTemperature, label='Johnson')
pres_Nixon = city_data[city_data["President"]=='Nixon']
ax.scatter(pres_Nixon.dt, pres_Nixon.AverageTemperature, label='Nixon')
pres_Ford = city_data[city_data["President"]=='Ford']
ax.scatter(pres_Ford.dt, pres_Ford.AverageTemperature, label='Ford')
pres_Carter = city_data[city_data["President"]=='Carter']
ax.scatter(pres_Carter.dt, pres_Carter.AverageTemperature, label='Carter')
pres_Reagan = city_data[city_data["President"]=='Reagan']
ax.scatter(pres_Reagan.dt, pres_Reagan.AverageTemperature, label='Reagan')
pres_Bush1 = city_data[city_data["President"]=='Bush1']
ax.scatter(pres_Bush1.dt, pres_Bush1.AverageTemperature, label='G.H.W. Bush')
pres_Clinton = city_data[city_data["President"]=='Clinton']
ax.scatter(pres_Clinton.dt, pres_Clinton.AverageTemperature, label='Clinton')
pres_Bush2 = city_data[city_data["President"]=='Bush2']
ax.scatter(pres_Bush2.dt, pres_Bush2.AverageTemperature, label='G.W. Bush')
pres_Obama = city_data[city_data["President"]=='Obama']
ax.scatter(pres_Obama.dt, pres_Obama.AverageTemperature, label='Obama')


ax.set_title('Average daily temperature for Specified City')
plt.legend(title='Presidents', bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP)
plt.savefig('cis4301_graph.png')
plt.show()
```
